```python
class Car:
  car_type = "Sedan"

  def __init__(self, name, mileage):
    self.name = name
    self.mileage = mileage

  def description(self):
    return f"The {self.name} car gives the mileage of {self.mileage}km/l"

  def car_max_speed(self, speed):
    return f"The {self.name} runs at the maximum speed of {speed} km/hr"

car1 = Car("Toyata", 2000)
car2 = Car("Honda", 3000)
print(car1.description())
print(car2.description())
```

```python
class animals:
  def __init__(self, name, habitat, temperature):
    self.name = name
    self.habitat = habitat
    self.temperature = temperature
  def feature(self):
    return f"The {self.name} which has a {self.temperature} body temperature and lives in {animals.habitat}"

mammals = animals("Dog", "Land", "Warm")
birds = animals("Sparrow", "Air", "Warm")
reptiles = animals("Snake", "Terrestrial", "Cold")
print(mammals.feature())
```

```python
import math

class shapes:
  def __init__(self, name, sides):
    self.name = name
    self.sides = sides

  def AreaSolver(self):
      if self.name == "Square":
        square_area = self.sides**2
        return square_area
      elif self.name == "Triangle":
       base = float(input("Enter the base length of the triangle: "))
       height = float(input("Enter the height of the triangle: "))
       triangle_area = 0.5 * base * height
       return triangle_area
      elif self.name  == "Rectangle":
       length = float(input("Enter the length of the rectangle: "))
       width = float(input("Enter the width of the rectangle: "))
       rectangle_area = length * width
       return rectangle_area
      elif self.name == "Circle":
       radius = float(input("Enter the radius of the circle: "))
       circle_area = math.pi * (radius**2)
       return circle_area

  def shape_description(self):
    return f"The {self.name} has {self.sides} side/s and has an area of {self.AreaSolver()}"

shape1 = shapes("Square", 5)
shape2 = shapes("Triangle", 3)
shape3 = shapes("Rectangle", 4)
shape4 = shapes("Circle", 0)
print(shape1.shape_description())
print(shape2.shape_description())
print(shape3.shape_description())
print(shape4.shape_description())
```
