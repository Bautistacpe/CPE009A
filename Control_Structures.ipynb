Elif Statements

```python
a=12
b=12
if b>a:
  print("b is greater than a")
elif a==b:
  print("a and b are equal")
```

Else Statements

```python
a=200
b=33
if b>a:
  print("b is greater than a")
elif a==b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

```python
a=2
b=330
print("A") if a>b else print("B")
```

And-both conditions are true

```python
a = 200
b = 33
c = 500
if a>b and c>a:
  print("Both conditions are True")
```

Or-at least one of the conditon is true

```python
a=200
b=33
c=500
if a>b or a>c:
  print("At least of of the conditions is True")
```
