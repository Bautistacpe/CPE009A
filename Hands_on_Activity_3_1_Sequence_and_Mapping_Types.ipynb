2. Simple Word Filter
*  For a given string input, replace all the words “stupid” with an asterisk * equal to the length of the string. The new
string value should be displayed with the asterisks.
3. Phonebook
*  Create a simple phonebook program that can read from a list of dictionaries. The program should be able to display a
person’s name, contact, and address based from a user input which is the id of the record.

```python
#1.
coordinates_list = [(1,1), (2,3)]
x1 = coordinates_list[0][0]
x2 = coordinates_list[1][0]
y1 = coordinates_list[0][1]
y2 = coordinates_list[1][1]
distance_coordinate = ((x2-x1)**2 + (y2-y1)**2)**1/2
print(float(distance_coordinate))
```

```python
#2.
sentence = input("Enter a sentence:")
dictionary = {"stupid","Stupid","STUPID"}
for word in dictionary:
  if word in sentence:
   sentence = sentence.replace(word,"*" * len(word))
   print(sentence)
```
