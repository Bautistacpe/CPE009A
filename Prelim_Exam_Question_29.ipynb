```python
counter = 0
grade = 0
total_grade = 0
student_name = input("Please enter your name: ")
while True:
 course_code = input("Enter a subject: ")
 if course_code != "quit":
  grade = int(input("Please enter your grade of the subject: "))
  print(" ")
  print("Subject:", course_code)
  print("Grade:", grade)
  if grade >= 90:
   print("Grade Equivalent: A+")
  elif 90 > grade >= 80:
   print("Grade Equivalent: A")
  elif 80 > grade >= 70:
   print("Grade Equivalent: B+")
  elif 70 > grade >= 60:
   print("Grade Equivalent: B")
  elif 60 > grade >= 50:
   print("Grade Equivalent: C")
  elif 50 > grade >= 40:
   print("Grade Equivalent: D")
  else:
   print("Grade Equivalent: F")
  counter +=1
  total_grade += grade
  print(" ")
 else:
  total_grade /= counter
  print("Student Name:", student_name)
  print("Total Grade:", total_grade, "%")
  if total_grade >= 90:
   print("Remarks: Outstanding")
  elif 90 > total_grade >= 80:
   print("Remarks: Excellent")
  elif 80 > total_grade >= 70:
   print("Remarks: Very Good")
  elif 70 > total_grade >= 60:
   print("Remarks: Good")
  elif 60 > total_grade >= 40:
   print("Remarks: Satisfactory")
  else:
   print("Remarks: Needs Imporvement")
  break
```
